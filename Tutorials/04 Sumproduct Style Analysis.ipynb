```python
def sumproduct(dataframe, column_1_label, column_2_label):
    column_1 = np.array(dataframe[column_1_label])
    column_2 = np.array(dataframe[column_2_label])
    line_totals = column_1 * column_2
    return line_totals.sum()
```

```python
sumproduct(sales_data, "PRICEEACH", "QUANTITYORDERED")
```
