```python
purchases = pd.read_csv("san_francisco_purchase_data.csv")
purchases.head()
```

```python
purchases["Vendor Name"].nunique()
```

```python
purchases['Post Date - Current'] = purchases['Post Date - Current'].datetime.datetime.strptime('2013-01-01 09:10:12', '%Y-%m-%d %H:%M:%S')
```

```python
purchases[(purchases['Post Date - Current'] > '01/01/2015') & (purchases['Post Date - Current'] <= '05/30/2015')]
```

```python
purchases_by_vendor = purchases.pivot_table(values=["Encumbered Amount"], index=["Vendor Name"], aggfunc=np.sum)
purchases_by_vendor.head()
```

```python
writer = pd.ExcelWriter('purchases_by_vendor.xlsx', engine='xlsxwriter')
purchases_by_vendor.to_excel(writer,"PbyV")
writer.save()
```
